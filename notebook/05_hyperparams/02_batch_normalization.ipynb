# Batch Normalization

```text
- This is the process of normalizing the layers' inputs by re-centering and re-scaling.
- The activation distributions can shift as they pass through the hidden layers which can lead to covariance shifts or lead to vanishing/exploding gradients.
- The solution is to normalize the inputs to each layer (batch normalization).
- Batch normalization can be implemented/applied in two ways:
  - it can be applied to the linear computation 'before' the activation function.
  - it can also be applied 'after' passing the input in an activation function. 
```

#### Preprocess Data


````python
# Select all the features bar the target.
X = data.iloc[:, :-1]

# Normalize the data
norm_ = Normalizer(min_value=0, max_value=1)
X_norm = norm_.fit_transform(X=X)

X_norm.head()
````

````python
# Binarize output
choice_list = [0, 1]

# 3, 4, and 5 : 0, else 1
data["target"] = np.where(data["quality"].isin(np.arange(3, 6)), choice_list[0], choice_list[1])

data[["quality", "target"]].sample(5, random_state=5)
````

````python
# Convert to Tensors
X = torch.Tensor(X_norm.to_numpy())

# Requred by PyTorch to be 2D instead of 1D
y = torch.Tensor(data["target"].to_numpy()).reshape(-1, 1)

# Split data
TEST_SIZE, RANDOM_STATE = 0.2, 123

X_train, X_validation, y_train, y_validation = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE
)

X_train.shape, X_validation.shape
````

````python
BATCH_SIZE, SHUFFLE = 16, True
DROP_LAST = True


# Create PyTorch datasets and dataloader objects.
train_data = TensorDataset(X_train, y_train)
validation_data = TensorDataset(X_validation, y_validation)

train_DL = DataLoader(
    dataset=train_data, batch_size=BATCH_SIZE, shuffle=SHUFFLE, drop_last=DROP_LAST
)
validation_DL = DataLoader(
    dataset=validation_data,
    batch_size=validation_data.tensors[0].shape[0],
)
````

````python
# Create a custom class that supports batch normalization.
class Net(nn.Module):
    """This is an ANN architecture that supports batch normalization."""

    def __init__(self) -> None:
        super().__init__()
        self.input = nn.Linear(11, 16)

        self.hidden_1 = nn.Linear(16, 32)
        # The input is the input_size of the prev. layer
        self.batch_norm_1 = nn.BatchNorm1d(16)
        self.hidden_2 = nn.Linear(32, 8)
        # The input is the input_size of the prev. layer
        self.batch_norm_2 = nn.BatchNorm1d(32)

        self.output = nn.Linear(8, 1)

    def forward(self, X: torch.Tensor, is_batch_norm: bool) -> torch.Tensor:
        """This is used to perform forward propagtion."""
        X = F.relu(self.input(X))

        if is_batch_norm:
            X = self.batch_norm_1(X)
            X = F.relu(self.hidden_1(X))
            X = self.batch_norm_2(X)
            X = F.relu(self.hidden_2(X))
        else:
            X = F.relu(self.hidden_1(X))
            X = F.relu(self.hidden_2(X))

        X = torch.sigmoid(self.output(X))
        return X


# Train ANN model
def train_model(
    *,
    train_dataloader: DataLoader,
    validation_dataloader: DataLoader,
    is_batch_norm: bool = True,
) -> tuple[float, float, float]:
    """This is used to train an ANN model with batch normalization."""
    LEARNING_RATE, EPOCHS = 0.01, 1_000
    THRESH, PCT = 0.5, 100

    net = Net()
    optimizer = torch.optim.SGD(params=net.parameters(), lr=LEARNING_RATE)
    criterion, train_loss = nn.BCELoss(), []
    train_accuracy, validation_accuracy = [], []

    for epoch_idx in np.arange(EPOCHS):
        net.train()
        batch_accuracy, batch_loss = [], []

        for X_, y_ in train_dataloader:
            # Reset gradients
            optimizer.zero_grad()

            # Compute forward prop and loss
            y_proba = net(X_, is_batch_norm)
            loss: torch.Tensor = criterion(y_proba, y_)

            # Back prop
            loss.backward()
            optimizer.step()

            # Batch metrics
            acc = torch.mean(((y_proba > THRESH) == y_).float()) * PCT
            batch_accuracy.append(acc.detach())
            batch_loss.append(loss.detach())

        # Train metrics
        train_accuracy.append(np.mean(batch_accuracy))
        train_loss.append(np.mean(batch_loss))

        # Validation metrics
        net.eval()
        
        X_val, y_val = next(iter(validation_dataloader))
        y_proba_val = net(X_val, is_batch_norm)
        acc = torch.mean(((y_proba_val > THRESH) == y_val).float()) * PCT
        validation_accuracy.append(acc.detach())

    return (train_accuracy, validation_accuracy, train_loss)
````

````python
# With batch normalization
train_accuracy, validation_accuracy, train_loss = train_model(
    train_dataloader=train_DL, validation_dataloader=validation_DL, is_batch_norm=True
)
````

````python
# Without batch normalization
train_accuracy_2, validation_accuracy_2, train_loss_2 = train_model(
    train_dataloader=train_DL, validation_dataloader=validation_DL, is_batch_norm=False
)
````

````python
# Visualize the results
fig, ax = plt.subplots(1, 3, figsize=(17, 5))

ax[0].plot(train_loss, label="With batchnorm")
ax[0].plot(train_loss_2, label="No batchnorm")
ax[0].set_title("Losses")

ax[1].plot(train_accuracy, label="With batchnorm")
ax[1].plot(train_accuracy_2, label="No batchnorm")
ax[1].set_title("Train accuracy")

ax[2].plot(validation_accuracy, label="With batchnorm")
ax[2].plot(validation_accuracy_2, label="No batchnorm")
ax[2].set_title("Test accuracy")

labels = ["With batchnorm", "No batchnorm"]
# Common features
for idx in range(3):
    ax[idx].legend(labels)
    ax[idx].set_xlabel("Epoch")
    ax[idx].grid()

    if idx > 0:
        ax[idx].set_ylabel("Accuracy (%)")
        ax[idx].set_ylim(70, 100)

plt.show()
````

### Exercise

```text
- The model training above implemented batch normalization `BEFORE` applying the activation function.
- The next goal is to implement batch normalization `AFTER` applying the activation function.
```

````python
class Net(nn.Module):
    """This is an ANN architecture that supports batch normalization.
    Batch normalization is applied after the activation function.
    """

    def __init__(self) -> None:
        super().__init__()
        self.input = nn.Linear(11, 16)

        self.hidden_1 = nn.Linear(16, 32)
        # The input is the output_size of the prev. layer
        # This is because batch normalization will be applied
        # after the activation function
        self.batch_norm_1 = nn.BatchNorm1d(32)
        self.hidden_2 = nn.Linear(32, 8)
        self.batch_norm_2 = nn.BatchNorm1d(8)

        self.output = nn.Linear(8, 1)

    def forward(self, X: torch.Tensor, is_batch_norm: bool) -> torch.Tensor:
        """This is used to perform forward propagtion."""
        X = F.relu(self.input(X))

        if is_batch_norm:
            # Apply normalization after the activation function
            X = F.relu(self.hidden_1(X))
            X = self.batch_norm_1(X)
            X = F.relu(self.hidden_2(X))
            X = self.batch_norm_2(X)
        else:
            X = F.relu(self.hidden_1(X))
            X = F.relu(self.hidden_2(X))

        X = torch.sigmoid(self.output(X))
        return X
````

````python
# With batch normalization
train_accuracy, validation_accuracy, train_loss = train_model(
    train_dataloader=train_DL, validation_dataloader=validation_DL, is_batch_norm=True
)

# Without batch normalization
train_accuracy_2, validation_accuracy_2, train_loss_2 = train_model(
    train_dataloader=train_DL, validation_dataloader=validation_DL, is_batch_norm=False
)
````

````python
# Visualize the results
fig, ax = plt.subplots(1, 3, figsize=(17, 5))

ax[0].plot(train_loss, label="With batchnorm")
ax[0].plot(train_loss_2, label="No batchnorm")
ax[0].set_title("Losses")

ax[1].plot(train_accuracy, label="With batchnorm")
ax[1].plot(train_accuracy_2, label="No batchnorm")
ax[1].set_title("Train accuracy")

ax[2].plot(validation_accuracy, label="With batchnorm")
ax[2].plot(validation_accuracy_2, label="No batchnorm")
ax[2].set_title("Test accuracy")

labels = ["With batchnorm", "No batchnorm"]
# Common features
for idx in range(3):
    ax[idx].legend(labels)
    ax[idx].set_xlabel("Epoch")
    ax[idx].grid()

    if idx > 0:
        ax[idx].set_ylabel("Accuracy (%)")
        ax[idx].set_ylim(70, 100)

plt.show()
````
